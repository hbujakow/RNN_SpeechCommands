```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
def plot_both(df1,df2,df3,file_name):
    mean_df = pd.concat([df1, df2, df3]).groupby('epoch').mean().reset_index()
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))

    axes[0].plot(mean_df['epoch'], mean_df['train_loss'], label='Train', color='blue', linewidth=2)
    axes[0].plot(mean_df['epoch'], mean_df['val_loss'], label='Val', color='orange', linewidth=2)
    axes[0].legend()

    axes[0].plot(df1['epoch'], df1['train_loss'], label='Train', color='blue', alpha=0.3)
    axes[0].plot(df1['epoch'], df1['val_loss'], label='Val', color='orange', alpha=0.3)
    axes[0].plot(df2['epoch'], df2['train_loss'], label='Train', color='blue', alpha=0.3)
    axes[0].plot(df2['epoch'], df2['val_loss'], label='Val', color='orange', alpha=0.3)
    axes[0].plot(df3['epoch'], df3['train_loss'], label='Train', color='blue', alpha=0.3)
    axes[0].plot(df3['epoch'], df3['val_loss'], label='Val', color='orange', alpha=0.3)


    axes[0].set_title("Loss")
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Loss')

    axes[1].plot(mean_df['epoch'], mean_df['train_accuracy'], label='Train', color='blue', linewidth=2)
    axes[1].plot(mean_df['epoch'], mean_df['val_accuracy'], label='Val', color='orange', linewidth=2)
    axes[1].legend()

    axes[1].plot(df1['epoch'], df1['train_accuracy'], label='Train', color='blue', alpha=0.3)
    axes[1].plot(df1['epoch'], df1['val_accuracy'], label='Val', color='orange', alpha=0.3)
    axes[1].plot(df2['epoch'], df2['train_accuracy'], label='Train', color='blue', alpha=0.3)
    axes[1].plot(df2['epoch'], df2['val_accuracy'], label='Val', color='orange', alpha=0.3)
    axes[1].plot(df3['epoch'], df3['train_accuracy'], label='Train', color='blue', alpha=0.3)
    axes[1].plot(df3['epoch'], df3['val_accuracy'], label='Val', color='orange', alpha=0.3)


    axes[1].set_title("Accuracy")
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Accuracy')
    axes[1].set_ylim(0, 1)
    # axes[1].ylim(0, 1)

    plt.tight_layout()
    plt.savefig(file_name)
    plt.show()
```

```python
import os

folders = os.listdir('LSTM/results')
folders = [folder.split('_SEED_')[0] for folder in folders if folder.startswith('GRU')]
```

```python
folders = list(set(folders))
folders
```

```python
for fol in folders:
    path = 'LSTM/results/'+fol
    df1 = pd.read_csv(f'LSTM/results/{fol}_SEED_1/results.csv')
    df2 = pd.read_csv(f'LSTM/results/{fol}_SEED_2/results.csv')
    df3 = pd.read_csv(f'LSTM/results/{fol}_SEED_3/results.csv')
    print(fol)
    plot_both(df1,df2,df3,"LSTM/results/"+fol+".png")
```
