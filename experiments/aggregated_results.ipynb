# RNN, LSTM, GRU 

```python
import os
import json
import pandas as pd


results = {
    'model': [],
    'num_layers': [],
    'hidden_size': [],
    'seed': [],
    'test_acc': [],
}

root = 'LSTM/results/'

for i, folder in enumerate(list(filter(lambda x: not x.endswith('.png'), sorted(os.listdir(root))))):
    model_name = folder.split('__')[0] 
    params = folder.split('__')[1]
    splitted = params.split('_')
    num_layers = splitted[2]
    hidden_size = splitted[4]
    seed = splitted[-1]

    with open(root + folder + '/test_results.json', 'r')  as f:
        test_acc = json.load(f)['test_accuracy']
    
    results['model'].append(model_name)
    results['num_layers'].append(int(num_layers))
    results['hidden_size'].append(int(hidden_size))
    results['seed'].append(int(seed))
    results['test_acc'].append(round(test_acc, 3))
```

```python
results = pd.DataFrame(results)
results.head()
```

```python
tmp = results.groupby(['model', 'num_layers', 'hidden_size'])['test_acc'].agg(['mean', 'std']).round(3)
tmp
```

# Whisper and AST

```python
import os
import json
import pandas as pd


results = {
    'model': [],
    'pretrained': [],
    'seed': [],
    'test_acc': [],
}

root = 'whisper/checkpoints/'

for if_pretrained in os.listdir(root):
    path = os.path.join(root, if_pretrained)
    for experiment in os.listdir(root + if_pretrained):
        if experiment.endswith('.png'):
            continue
        experiment_path = os.path.join(path, experiment)
        seed = int(experiment_path[-1])
        with open(os.path.join(experiment_path, 'test_metrics.json'), 'r')  as f:
            test_acc = json.load(f)['test_accuracy']
        results['model'].append('whisper')
        results['pretrained'].append(if_pretrained)
        results['seed'].append(seed)
        results['test_acc'].append(round(test_acc, 3))
results1 = pd.DataFrame(results)
```

```python
results = {
    'model': [],
    'pretrained': [],
    'seed': [],
    'test_acc': [],
}

root = 'AST/checkpoints/'

for if_pretrained in os.listdir(root):
    path = os.path.join(root, if_pretrained)
    for experiment in os.listdir(root + if_pretrained):
        if experiment.endswith('.png'):
            continue
        experiment_path = os.path.join(path, experiment)
        seed = int(experiment_path[-1])
        with open(os.path.join(experiment_path, 'test_metrics.json'), 'r')  as f:
            test_acc = json.load(f)['test_accuracy']
        results['model'].append('ast')
        results['pretrained'].append(if_pretrained)
        results['seed'].append(seed)
        results['test_acc'].append(round(test_acc, 3))
results2 = pd.DataFrame(results)
```

```python
tmp = results.groupby(['model', 'pretrained'])['test_acc'].agg(['mean', 'std']).round(3)
tmp
```
